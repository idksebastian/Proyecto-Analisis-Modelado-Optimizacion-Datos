```python
#Histograma del salario mensual promedio
plt.figure(figsize=(8,5))
sns.histplot(df['SALARIO MENSUAL PROMEDIO'], bins=30)
plt.title('Distribución del Salario Mensual Promedio')
plt.xlabel('Salario mensual promedio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Boxplot del salario
plt.figure(figsize=(8,4))
sns.boxplot(x=df['SALARIO MENSUAL PROMEDIO'])
plt.title('Boxplot del Salario Mensual Promedio')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='ORDEN',
    y='SALARIO MENSUAL PROMEDIO'
)

plt.yscale('log')  # Escala logarítmica
plt.title('Salario Mensual Promedio según Orden de la Entidad (escala log)')
plt.xlabel('Orden de la Entidad')
plt.ylabel('Salario Mensual Promedio (log)')
plt.show()
```

```python
genero_cols = ['GENERO: HOMBRE', 'GENERO: MUJER']

df[genero_cols].sum().plot(
    kind='bar',
    figsize=(6,4),
    title='Distribución total por género'
)
plt.ylabel('Cantidad de empleados')
plt.show()
```

```python
educacion_cols = [
    'NIVEL EDUCATIVO: PRIMARIA',
    'NIVEL EDUCATIVO: SECUNDARIA',
    'NIVEL EDUCATIVO: TÉCNICA-TECNOLÓGICA',
    'NIVEL EDUCATIVO: PROFESIONAL',
    'NIVEL EDUCATIVO: ESPECIALIZACIÓN',
    'NIVEL EDUCATIVO: MAESTRÍA',
    'NIVEL EDUCATIVO: DOCTORADO',
    'NIVEL EDUCATIVO: SIN DATOS'
]

# Convert relevant columns to numeric, coercing errors to NaN
for col in educacion_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

df[educacion_cols].sum().plot(
    kind='bar',
    figsize=(9,4),
    title='Distribución por Nivel Educativo'
)
plt.ylabel('Cantidad de empleados')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(
    df.select_dtypes(include='number').corr(),
    cmap='coolwarm',
    center=0
)
plt.title('Matriz de Correlación (Datos en Bruto)')
plt.show()
```
