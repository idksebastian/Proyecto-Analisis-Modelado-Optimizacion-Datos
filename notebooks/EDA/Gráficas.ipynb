```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("empleo_publico_limpio.csv")

df.head()
```

```python
cols_exp = [
    'EXPERIENCIA LABORAL: MENOR A 1 AÑO',
    'EXPERIENCIA LABORAL: DE 1 A 10 AÑOS',
    'EXPERIENCIA LABORAL: DE 11 A 20 AÑOS',
    'EXPERIENCIA LABORAL: DE 21 A 30 AÑOS',
    'EXPERIENCIA LABORAL: DE 31 A 40 AÑOS',
    'EXPERIENCIA LABORAL: MAYOR A 40 AÑOS'
]

for col in cols_exp:
    df[col] = pd.to_numeric(df[col], errors="coerce")

plt.figure()
df[cols_exp].sum().plot(kind="bar")

plt.title("Distribución por experiencia laboral")
plt.xlabel("Experiencia laboral")
plt.ylabel("Cantidad")
plt.show()
```

```python
cols_edad = [
    'EDAD: MENOR O IGUAL A 29',
    'EDAD: 30-39',
    'EDAD: 40-49',
    'EDAD: 50-62',
    'EDAD: 63 Y MAS'
]

for col in cols_edad:
    df[col] = pd.to_numeric(df[col], errors="coerce")

plt.figure()
df[cols_edad].sum().plot(kind="bar")

plt.title("Distribución por rangos de edad")
plt.xlabel("Rango de edad")
plt.ylabel("Cantidad")
plt.show()
```

```python
cols_nivel_edu = [
    'NIVEL EDUCATIVO: PRIMARIA',
    'NIVEL EDUCATIVO: SECUNDARIA',
    'NIVEL EDUCATIVO: TÉCNICA-TECNOLÓGICA',
    'NIVEL EDUCATIVO: PROFESIONAL',
    'NIVEL EDUCATIVO: ESPECIALIZACIÓN',
    'NIVEL EDUCATIVO: MAESTRÍA',
    'NIVEL EDUCATIVO: DOCTORADO'
]

for col in cols_nivel_edu:
    df[col] = pd.to_numeric(df[col], errors="coerce")

plt.figure()
df[cols_nivel_edu].sum().plot(kind="bar")

plt.title("Distribución por nivel educativo")
plt.xlabel("Nivel educativo")
plt.ylabel("Cantidad")
plt.show()
```

```python
plt.figure()

genero = df[["GENERO: HOMBRE", "GENERO: MUJER"]].sum()
genero.plot(kind="bar")

plt.title("Distribución de empleados por género")
plt.xlabel("Género")
plt.ylabel("Cantidad")
plt.show()
```

```python
plt.figure()
df["NIVEL"].value_counts().plot(kind="bar")

plt.title("Distribución por nivel organizacional")
plt.xlabel("Nivel")
plt.ylabel("Cantidad de registros")
plt.show()
```

```python
plt.figure()
df["TIPO DE VINCULACIÓN"].value_counts().plot(kind="bar")

plt.title("Distribución por tipo de vinculación")
plt.xlabel("Tipo de vinculación")
plt.ylabel("Cantidad de registros")
plt.show()
```

```python
plt.figure()

df.groupby("SECTOR")["SALARIO MENSUAL PROMEDIO"] \
  .mean() \
  .sort_values(ascending=False) \
  .head(10) \
  .plot(kind="bar")

plt.title("Top 10 sectores con mayor salario promedio")
plt.xlabel("Sector")
plt.ylabel("Salario promedio")
plt.show()
```

```python
plt.figure()
plt.hist(df["SALARIO MENSUAL PROMEDIO"], bins=30)
plt.title("Distribución del Salario Mensual Promedio")
plt.xlabel("Salario")
plt.ylabel("Frecuencia")
plt.show()
```
